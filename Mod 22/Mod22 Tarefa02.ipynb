````python
import pandas as pd
import plotly.express as px
from plotly.graph_objs import Layout
from plotly import graph_objects
````

````python
df = pd.read_csv('dados_historicos_acoes.csv', header=[0,1], index_col=0)
````

````python
df = df.dropna()
df
````

#### 2. Média móvel
1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
df_std = df.rolling(90).std()
df_std
````

````python
df_mw = df.rolling(window=90).mean()
df_mw.std()
````

````python
df_media_movel = df_mw.stack().reset_index()
df_media_movel.rename(columns={'Date' : 'Data' , 'Price' : 'Ativo'}, inplace=True)
df_media_movel
````

````python
fig = px.line(data_frame=df_media_movel,
              x='Data',
              y='Close', 
              color='Ticker',
              template='plotly_dark')
fig.show()
````

````python
df_std_adjclose = df_std.stack().reset_index()
df_std_adjclose.rename(columns={'Date' : 'Data' , 'Price' : 'Ativo'}, inplace=True)
df_std_adjclose
````

````python
df_std_adjclose = df_std.reset_index()
df_std_adjclose.columns = ['Date'] + [f"{col[0]}_{col[1]}" for col in df_std_adjclose.columns[1:]]
````

````python
df_std_adjclose = df_std_adjclose.melt(id_vars=['Date'], var_name='ativo', value_name='valor')
````

````python
ativos_desejados = ['Close_MGLU3.SA', 'Close_MSFT', 'Close_NVDA']
df_filtrado = df_std_adjclose[df_std_adjclose['ativo'].isin(ativos_desejados)]

fig = px.line(data_frame=df_filtrado,
              x='Date', 
              y='valor',
              color='ativo',
              template='plotly_dark')

fig.show()
````

Os gráficos mostram que os períodos de maior volatilidade (desvio padrão) coincidem com quedas ou oscilações nos preços médios dos ativos. MSFT foi o mais volátil, com picos em 2022, enquanto NVDA apresentou comportamento mais estável ao longo do tempo. A análise indica que a volatilidade acompanha momentos de maior instabilidade nos preços, reforçando a relação entre os dois indicadores.

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python
df_60 = df.xs(key='MSFT', level=1, axis=1).tail(n=60)
df_60.head()
````

````python
layout = Layout(paper_bgcolor='rgba(0,0,0,0)',
                plot_bgcolor='black',
                height=600)

graph = {
    'x': df_60.index,
    'open': df_60.Open,
    'close': df_60.Close,
    'high': df_60.High,
    'low': df_60.Low,
    'type': 'candlestick',
    'name': 'MSFT',
    'showlegend': True
}

fig = graph_objects.Figure(data=[graph], layout_title='', layout=layout)

fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

fig
````
