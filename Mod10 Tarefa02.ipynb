````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv('Previsao.csv')
df
````

````python
ax = sns.pointplot(x= 'posse_de_veiculo', y= 'renda', data= df, dodge= True)
````

````python
sns.pointplot(x= 'posse_de_imovel', y= 'renda', data= df, dodge= True)
````

````python
sns.pointplot(x= 'posse_de_veiculo', y= 'renda', hue= 'posse_de_imovel',
                  data= df, dodge= True)
````

````python
sns.pointplot(x= 'posse_de_imovel', y= 'renda', hue= 'posse_de_veiculo',
                  data= df, dodge= True)
````

Existem diferenças significantes na renda entre grupos com e sem posse de veiculo, assim como para grupos com e sem posse de imovel.
