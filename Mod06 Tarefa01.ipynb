```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('SINASC_RO_2019.csv')
df.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
df['GESTACAO'] = df['GESTACAO'].str.replace('semanas', '')
```

```python
fig, axs = plt.subplots(3, 2, figsize = (15, 10))

sns.countplot(ax = axs [0, 0], x= 'APGAR1', data= df)

sns.countplot(ax = axs [0, 1], x= 'APGAR5', data= df)

sns.countplot(ax = axs [1, 0], x= 'SEXO', data= df)

sns.countplot(ax = axs [1, 1], x= 'GESTACAO', data= df)

sns.countplot(ax = axs [2, 0], x= 'ESTCIVMAE', data= df)

fig.delaxes(axs[2,1])
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
sns.barplot(x= 'CONSULTAS', y= 'APGAR5', data= df)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axs = plt.subplots(2, 2, figsize= (20, 10))

sns.countplot(ax= axs [0, 0], x= 'SEXO', data= df)

sns.countplot(ax= axs [0, 1], x= 'GESTACAO', data= df)

sns.countplot(ax= axs [1, 0], x= 'ESTCIVMAE', data= df)

sns.countplot(ax= axs [1, 1], x= 'CONSULTAS', data= df)
```
